## If Statement
- If block will be executed only when the condition is true

```python
isItRaining = True
if isItRaining:
  print('Please take an umbrella')
```

```python
weather = input('How is the weather today? ')
if weather == 'rainy':
  print('Please take an umbrella')
```

## If Else

```python
weather = input('How is the weather today? ')
if weather == 'rainy':
  print('Please stay at home')
else:
  print('Enjoy your day')
```

```python
number = int(input("Please provide any number, I'll determine it is a positive or negative number"))
if number>0:
  print('Given Number is positive')
elif number==0:
  print('Given Number is zero')
else:
  print('Given Number is negative')
```

```python
num1 = int(input('Please provide first number'))
num2 = int(input('Please provide second number'))
if num1>num2:
  print(f'{num1} is greater than {num2}')
else:
  print(f'{num2} is greater than {num1}')
```

```python
num = int(input('Please provide a number'))
if num%2==0:
  print(f'{num} is an even number')
else:
  print(f'{num} is an odd number')
```

```python
amount = int(input('Please Enter the amount you want to withdraww'))

if (amount % 500 == 0) or (amount % 200 == 0):
  print('Here is the cash')
else:
  print(f'Requested {amount} is not available')
```

```python
color = input('Please provide the color name: ')

if color == 'red':
  print('stop')
else:
  if color == 'yellow':
    print('ready')
  else:
    if color == 'green':
      print('go')
    else:
      print('invalid color')
```

## Elif

```python
num = int(input('Please provide a number: '))
if num>0:
  print('Given Number is positive')
elif num==0:
  print('Given Number is zero')
else:
  print('Given Number is negative')
```

```python
n = int(input())

if n%3 == 0:
  print('Fizz')
elif n%5 ==0:
  print('Fuzz')
elif (n%3 == 0) and (n%5==0):
  print('FizFuzz')
else:
  print(f"Criteria for the given number {n} doesn't match")
```
